#Pemilihan Fitur

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['SpeciesEncoded'] = le.fit_transform(df['Species'])
```

```python
#Fitur untuk clustering
X = df.drop(["Species", "SpeciesEncoded"], axis=1)
```

#Menentukan Jumlah Cluster (K) dengan Elbow Method

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia_list = []
K_range = range(1, 10)

for k in K_range:
  km = KMeans(n_clusters=k, random_state=42)
  km.fit(X)
  inertia_list.append(km.inertia_)

plt.figure(figsize=(7,5))
plt.plot(K_range, inertia_list, marker='o')
plt.title('Elbow Method untuk Memnentukan jumlah Cluster')
plt.xlabel('Jumlah Cluster (K)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

#4. Membangun Model K-Means

```python
from sklearn.cluster import KMeans
km = KMeans(
    n_clusters=3,
    init="k-means++",
    random_state=42
)
```

```python
df["Cluster"] = km.fit_predict(X)
print("Cluster ditemukan:", df["Cluster"].unique())
```

#5. Evaluasi Kualitas Clustering

```python
from sklearn.metrics import silhouette_score

sil_score = silhouette_score(X, df["Cluster"])
print("\n===== SILHOUTTE SCORE =====")
print("Silhouette Score:", sil_score)
```

#6. Confusion Matrix (Perbandingan dengan Label Asli)

```python
from sklearn.metrics import confusion_matrix, classification_report
print("\n==== CONFUSION MATRIX =====")
cm = confusion_matrix(df["SpeciesEncoded"], df["Cluster"])
print(cm)
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(
    cm, annot=True, fmt='d', cmap='Blues',
    xticklabels=["Cluster 0", "Cluster 1", "Cluster 2"],
    yticklabels=["Cluster 0", "Cluster 1", "Cluster 2"]
)
plt.title("Confusion Matrix: k-Means vs Label Asli")
plt.xlabel("Cluster Prediksi")
plt.ylabel("Label Asli")
plt.show()

```

#2D dan 3D Visualisasi Hasil Model K-Means

```python
plt.figure(figsize=(7,5))
sns.scatterplot(
    x=df["SepalLengthCm"],
    y=df["SepalWidthCm"],
    hue=df["Cluster"],
    palette="deep"
)
plt.title("Visualisasi K-Means Clustering (2D)")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

sctter = ax.scatter(
    df["SepalLengthCm"],
    df["PetalLengthCm"],
    df["PetalWidthCm"],
    c=df["Cluster"],
    s=60
)

ax.set_title("3D plot K-Means Clustering pada Iris")
ax.set_xlabel("Sepal Length (cm)")
ax.set_ylabel("Petal Length (cm)")
ax.set_zlabel("Petal Width (cm)")

# Legend Cluster
legend_labels = [f"Cluster (i)" for i in range(3)]
ax.legend(handles=sctter.legend_elements()[0], labels=legend_labels)

plt.show()
```
